```python
import wave

def get_wav_duration(file_path):
    with wave.open(file_path, 'rb') as wf:
        # Get the number of frames and the frame rate
        num_frames = wf.getnframes()
        sampling_rate = wf.getframerate()

        # Calculate the duration in seconds
        duration = num_frames / float(sampling_rate)

        return duration
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

transcriptions_processed_ordering = []


model_enc = SentenceTransformer('all-MiniLM-L6-v2').to(device)
#transcriptions maps to audio file by index
encodings = model_enc.encode(transcriptions) #n,dim -> returns numpy array
print('encodings generic', encodings.shape)
average_encoding = np.mean(encodings, axis=0)
std_encoding = np.std(encodings, axis=0)

epsilon = 1e-9
encodings_standardised = np.array([np.abs(x - average_encoding) / (epsilon + std_encoding) for x in encodings]) #variance not necessary
# print('standardised', encodings_standardised.shape)
encodings_standardised = np.square(encodings_standardised)
# print('squared', encodings_standardised)
encodings_processed = np.sum(encodings_standardised, axis=1)
# print('SUM', encodings_processed.shape, encodings_processed)
encodings_processed_ordering = np.argsort(encodings_processed)[::-1]
# print(encodings_processed_ordering, encodings_processed_ordering.shape)
encodings_processed = [np.sqrt(encodings_standardised[index]) for index in encodings_processed_ordering]
filenames_processed_ordering = [idx_to_file[index] for index in encodings_processed_ordering]
transcriptions_processed_ordering = [transcriptions[index] for index in encodings_processed_ordering]
raw_encodings = [encodings[index] for index in encodings_processed_ordering]

print(encodings_processed_ordering)
print(f"Process the audio files as follows: {' ||| '.join(filenames_processed_ordering)}")
```

```python
import json
import pyodbc
import os
import sys
import ast

conf = None
with open(os.path.join(root, 'creds.json'), 'r') as file:
    # Load the JSON data into a Python dictionary
    conf = json.load(file)

def get_database_client(conf):
    """This function instanciates a client connection to our mongo server
    and it will crash otherwise"""

    try:
        connectionstring = 'Driver={};Server={};Database={};Trusted_Connection=yes;'
        connectionstring = connectionstring.format("SQL Server Native Client 11.0",
                                                    conf['SERVER_HOSTNAME'],
                                                   conf["DATABASE"])
        cnxn = pyodbc.connect(connectionstring)
    except Exception as e:
        error_message = "Error connecting to mongo instance by either :{0}"+ \
                        "\tHostname:{1}{0}\tDatabase:{2}{0}FULL Error:{3}"
        sys.stderr(error_message.format(os.linesep,
                                         conf['SERVER_HOSTNAME'],
                                         conf['DATABASE'],
                                        str(e)))
        exit()
    return cnxn

def create_table(cnxn):
    if checkTableExists(cnxn,conf['TABLE_NAME']) == False:
        ct = "".join(['CREATE TABLE dbo.', conf['TABLE_NAME'],
                     ' audio_file varchar(1000),',
                     ' transcription varchar(40000),',
                     ])
        cnxn.execute(ct)
        cnxn.commit()

def checkTableExists(dbcon, tablename):
      dbcur = dbcon.cursor() #is info_schema auto instantiated?
      dbcur.execute("""
          SELECT COUNT(*)
          FROM information_schema.tables
          WHERE table_name = '{0}'
          """.format(tablename.replace('\'', '\'\'')))
      if dbcur.fetchone()[0] == 1:
          dbcur.close()
          return True
      dbcur.close()
      return False

def write_record_email(cnxn,transcriptions=transcriptions_processed_ordering): #Did you mean LOG table here?
    # print(record)
    #transcriptions_processed_ordering
    for idx, text in enumerate(transcriptions):
        obj = ast.literal_eval(obj)

        insertStatement = "".join([f"INSERT INTO dbo.{conf['TABLE_NAME']}",
                                  "(audio_file,transcription) ",
                                  "VALUES ('"])
        insertStatement = insertStatement +"','".join([filenames_processed_ordering[idx],
                                                       text.replace("'", "''"),
                                                       ])+"')"
        try:
            cnxn.execute(insertStatement)
        except Exception as e:
            # print(record)
            # print(insertStatement)
            print(e)
            continue
            raise e
    cnxn.commit()

client = get_database_client(conf)
create_table(client)
write_record_email(client)
client.close()
```
